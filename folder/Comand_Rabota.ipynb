```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Пояснение данных

Name: Name of cereal\
mfr: Производитель крупы:\
A = American Home Food Products;\
G = General Mills\
K = Kelloggs\
N = Nabisco\
P = Post\
Q = Quaker Oats\
R = Ralston Purina\
type:\
cold\
hot\
calories: калорий на порцию\
protein: граммов белка\
fat: граммов жира\
sodium: миллиграммы натрия\
fiber: граммы пищевых волокон\
carbo: граммы сложных углеводов\
sugars: граммы сахаров\
potass: миллиграммы калия\
vitamins: витамины и минералы - 0, 25 или 100, что указывает на типичный процент, рекомендованный FDA\
shelf: полка-витрина (1, 2 или 3, считая от пола)\
weight: вес в унциях одной порции\
cups: количество чашек в одной порции\
рейтинг: рейтинг зерновых (возможно, из Consumer Reports?)

### График 10 самых лучших хлопьев

```python
data2 = data[["name","rating"]].value_counts().head(10).reset_index().sort_values(by="rating",ascending=False)
```

```python
plt.figure(figsize = (22,8))
plt.bar(x = data2["name"],height = data2["rating"])
plt.title("Топ 10 самых лучших хлопьев")
```

### Heatmap взаимоотношения одного элемента к другим

```python
numeric_columns = data.select_dtypes(include=['number'])

correlation_matrix = numeric_columns.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

### Зависимость калорийности от количества сахара в злаках

```python
data = pd.read_csv('cereal.csv') 
sns.scatterplot(x='sugars', y='calories', data=data, hue='mfr')
plt.title('Зависимость калорийности от количества сахара в злаках')
plt.xlabel('Количество сахара (г)')
plt.ylabel('Калорийность (ккал)')

plt.show()
```

### Сравнение каждого элемента с другими элементами

```python
sns.pairplot(data, hue="mfr", diag_kind="kde", plot_kws={'alpha':0.6, 's':50})
plt.suptitle('Сравнение', y=1.02)
plt.show()
```

### Значения калорий на каждой полке

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="shelf", y="calories", data=data)
plt.title("Значения калорий на каждой полке")
plt.xlabel("Shelf")
plt.ylabel("Calories")
plt.show()
```
