### Метод извлекающий текст из PDF файла

```python
def extract_text_from_pdf(pdf):
    doc = fitz.open(pdf)
    text = []
    for current_page in range(len(doc)):
        page = doc.load_page(current_page)
        page_text = page.get_text("text")
        text.append(page_text)
    return text
```

```python
import re

def pars_pdf(text):
    text2 = text.split('\n')
    raiting = 0
    name = ""
    activity = ""
    date_publish = ""
    used_indices = set()  

    # Поиск рейтинга
    for i, line in enumerate(text2):
        match_raiting = re.match(r'^\d+\.?\d*$', line.strip())
        if match_raiting:
            raiting = float(match_raiting.group())
            used_indices.add(i)
            # Проверяем строки после рейтинга
            if i + 1 < len(text2) and "Рейтинг" in text2[i + 1]:
                used_indices.add(i + 1)
                # Извлекаем название компании
                if i + 2 < len(text2) and i + 2 not in used_indices:
                    name = text2[i + 2].strip()
                    used_indices.add(i + 2)
                # Извлекаем деятельность
                if i + 3 < len(text2) and i + 3 not in used_indices:
                    activity = text2[i + 3].strip()
                    used_indices.add(i + 3)
                # Извлекаем дату публикации
                if i + 4 < len(text2) and i + 4 not in used_indices:
                    date_match = re.search(r'(\d+ час(?:а|ов)? назад|\d+ мин(?:ут(?:ы|а)?)? назад|\d+ мар(?:та)? в \d{1,2}:\d{2})', text2[i + 4])
                    if date_match:
                        date_publish = date_match.group()
                        used_indices.add(i + 4)
            break

    if not raiting:
        for keyword in ["Простой", "Средний"]:
            if keyword in text2:
                idx = text2.index(keyword)
                # Ищем дату публикации перед ключевым словом
                if idx > 1 and idx - 2 not in used_indices:
                    date_match = re.search(r'(\d+ час(?:а|ов)? назад|\d+ мин(?:ут(?:ы|а)?)? назад|\d+ мар(?:та)? в \d{1,2}:\d{2})', text2[idx - 2])
                    if date_match:
                        date_publish = date_match.group()
                        used_indices.update({idx - 2, idx})

    if not name:
        for i, line in enumerate(text2):
            if re.match(r'^[A-Za-zА-Яа-я0-9]+(?:[-_]?[A-Za-zА-Яа-я0-9]+)*$', line.strip()) and i not in used_indices:
                name = line.strip()
                used_indices.add(i)
                break

    if not activity:
        for i, line in enumerate(text2):
            if ("#" in line or "№" in line) and i not in used_indices:
                activity = line.strip()
                used_indices.add(i)
                break

    if not name and "Добавить компанию" in text2:
        idx = text2.index("Добавить компанию")
        if idx + 1 < len(text2) and idx + 1 not in used_indices:
            name = text2[idx + 1].strip()
            used_indices.update({idx, idx + 1})

    if not activity and "Пишет в хабы:" in text2:
        idx = text2.index("Пишет в хабы:")
        if idx > 0 and idx - 1 not in used_indices:
            activity = text2[idx - 1].strip()
            used_indices.update({idx - 1, idx})

    if not activity and "Привет, Хабр!" in text2:
        idx = text2.index("Привет, Хабр!")
        if idx + 1 < len(text2) and idx + 1 not in used_indices:
            activity = text2[idx + 1].strip()
            used_indices.update({idx, idx + 1})
   
    if not date_publish:
        for i, line in enumerate(text2):
            date_match = re.search(r'(\d+ час(?:а|ов)? назад|\d+ мин(?:ут(?:ы|а)?)? назад|\d+ мар(?:та)? в \d{1,2}:\d{2})', line)
            if date_match and i not in used_indices:
                date_publish = date_match.group()
                used_indices.add(i)
                break

    return name, raiting, activity, date_publish, used_indices
```

### Проделываем все это с остальными PDF файлами

```python
Brak = []
NameCompany = []
Description = []
Raiting = []
DataPublish = []
Activity = []
TextArticle = []

for pdf in all_pdf:
    try:
        x = extract_text_from_pdf(pdf)
        print(pdf.split('\\')[-1])
        N, R, A, D, used_indices = pars_pdf(x[0])  # Получаем used_indices
        T = ''.join([line for idx, line in enumerate(x[0].split('\n')) if idx not in used_indices])  # Исключаем использованные строки
        NameCompany.append(N)
        Raiting.append(R)
        DataPublish.append(D)
        Activity.append(A)
        TextArticle.append(T)
    except Exception as e:
        Brak.append(pdf)
```
